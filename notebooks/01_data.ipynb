```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from src.data import split_by_fold, DIAGNOSTIC_SUPERCLASSES
X = np.load('data/X_100.npy'); y = np.load('data/y_100.npy')
meta = pd.read_csv('data/ptb-xl/ptbxl_database.csv', index_col='ecg_id')
tr, va, te = split_by_fold(meta)
print('X', X.shape, '| y', y.shape)
print('class positives:', dict(zip(DIAGNOSTIC_SUPERCLASSES, y.sum(0).astype(int))))
print('train/val/test:', int(tr.sum()), int(va.sum()), int(te.sum()))
```

```python
fig, axes = plt.subplots(12, 1, figsize=(10, 11), sharex=True)
for i, ax in enumerate(axes):
    ax.plot(X[0, i], lw=.7); ax.set_ylabel(f'L{i+1}', fontsize=7); ax.set_yticks([])
axes[-1].set_xlabel('sample (10 s @ 100 Hz)')
fig.suptitle('Example PTB-XL 12-lead record'); fig.tight_layout(); plt.show()
```
