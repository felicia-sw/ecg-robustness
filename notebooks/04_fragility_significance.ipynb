```python
import pandas as pd
from src.analysis import significance, MODEL_ORDER
grid = pd.read_csv('results/grid.csv')
models = [m for m in MODEL_ORDER if m in set(grid.model)]
(fstat, fp), wh = significance(grid, models)
print(f'Friedman chi2 = {fstat:.2f}, p = {fp:.2e}')
for w in wh:
    print(f"  {w['pair'][0]:14s} vs {w['pair'][1]:14s} p_holm={w['p_holm']:.4f} "
          f"{'significant' if w['reject'] else 'ns'}")
```
