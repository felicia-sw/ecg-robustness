```python
import numpy as np, matplotlib.pyplot as plt
from src.corruptions import CORRUPTIONS, apply_corruption, SNR_DB_BY_SEVERITY
X = np.load('data/X_100.npy'); rec = X[0:1]; lead = 1
fig, axes = plt.subplots(len(CORRUPTIONS), 1, figsize=(11, 13), sharex=True)
for ax, name in zip(axes, CORRUPTIONS):
    ax.plot(rec[0, lead], 'k', lw=.8, label='clean')
    for s in [3, 5]:
        ax.plot(apply_corruption(rec, name, s)[0, lead], lw=.7, alpha=.7, label=f'sev {s}')
    ax.set_title(name, fontsize=9); ax.set_yticks([]); ax.legend(fontsize=6, ncol=3, loc='upper right')
fig.suptitle('ECG-C corruptions (lead II)'); fig.tight_layout(); plt.show()
```

```python
def measured_snr(x, xc):
    n = xc - x; ps = np.mean(x**2, -1); pn = np.mean(n**2, -1); m = pn > 0
    return float(np.mean(10*np.log10(ps[m]/pn[m])))
s = X[:200]
for name in ['baseline_wander', 'muscle_artifact', 'gaussian']:
    meas = {sev: round(measured_snr(s, apply_corruption(s, name, sev)), 2) for sev in [1, 3, 5]}
    tgt = {sev: SNR_DB_BY_SEVERITY[sev] for sev in [1, 3, 5]}
    print(f'{name:16s} measured {meas}  target {tgt}')
```
